```python
# creating player market information(player_mrk)
player_mrkt = pd.DataFrame(df_player, columns=['Name', 'Wage','Value', 'Release Clause'])
player_mrkt.head()
```

```python
def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0.0

player_mrkt['Wage'] = player_mrkt['Wage'].replace('[\€,]', '', regex=True).apply(value_to_float)
player_mrkt['Value'] = player_mrkt['Value'].replace('[\€,]', '', regex=True).apply(value_to_float)
player_mrkt['Release Clause'] = player_mrkt['Release Clause'].replace('[\€,]', '', regex=True).apply(value_to_float)

```

```python
print(player_mrkt.head())
print(player_mrkt.shape)
```

```python
#creating player other market and player personal information(player_other_info)
player_other_info = pd.DataFrame(df_player, columns=['ID', 'Name', 'Nationality','Club'])
player_other_info.head()
```

```python
# Merging the player_market_information(player_mrk) and player other information(player_other_informations)
player_info = pd.merge(player_other_info, player_mrkt, how='left')
player_info.head()
```
